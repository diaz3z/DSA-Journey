### Inverted Star pattern

```python
num = 5
```

```python
for i in range(num):
    for j in range(i + 1):
        print(" ", end="")
    for j in range(i, num-1):
        print("*", end="")
    for j in range(i, num):
        print("*", end="")
    print()
```

```python
for i in range(num):
    for j in range(i + 1):
        print("*", end=" ")
    print()
```

```python
for i in range(num):
    for j in range(i, num):
        print(" ", end=" ")
    for j in range(i + 1):
        print("*", end=" ")
    # for j in range(i, num):
    #     print("*", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
for i in range(num):
    for j in range(i + 1):
        if i % 2 == 0:
            print("A", end=" ")
        else:
            print("B", end=" ")
    print()
```

```python
n = 5
p = 1
for i in range(n):
    for j in range(i + 1):
        print(p, end=" ")
        p += 1
    print()
```

```python
row = 5
col = 10
for i in range(1, row+1):
    for j in range(1, col+1):
        if i == 1 or i == row or j == 1 or j == col:
            print("*", end=" ")
        else:
            print(" ",end=" ")
    print()
```
